```python
def findOnesCube(arr1, arr2):
# This function determs from the input of two 0's arrays all of the one's cubes
    onescube = []
    for i in range(len(arr2)):
        for j in range(len(arr1)): 
            first = arr2[i]
            second = arr1[j]
            diff = first[0] - second[0]
            if (diff == 1 or diff == 2 or diff == 4 or diff == 8 or diff == 16):
                onecube = [first[0], second [0], diff, False]
                onescube.append(onecube)
                arr2[i][1] = True
                arr1[j][1] = True
    return onescube,arr1,arr2
```

```python
def findTwosCube(arr1, arr2):
#This function determines from the input of two ones cubes arrays if there exists two cubes
    twoscubes=[]
    for i in range(len(arr2)):
        for j in range(len(arr1)):
            first = arr2[i]
            second = arr1[j]
            if first[2] == second[2] and ((first[3] == False) and (second[3] == False)) :
                diff2 = first[0] - second [0]
                twocube = [first[0], first[1], second[0], second[1], first[2] , diff2, False]
                twoscubes.append(twocube)
                arr2[i][3] = True
                arr1[j][3] = True
    return twoscubes,arr1,arr2
    
    
```

```python
def addBool (arr):
# This function adds a bool to the end of a list 
    new_arr = []
    for i in range(len(arr)):
        if arr[i] != [] :
            x = [arr[i], False]
            new_arr.append(x)
    return new_arr
```

```python
def findUnuseds(arr):
#This function finds the unused PIs
    temp_arr = arr
    new_arr = []
    length = len(arr)
    
    # If a bool is set to true that means that the cubes have been used
    for i in range(length):
        if temp_arr[i][-1] == True:
            arr[i] = []
        #If a cube has not been used, it is time to wipe the bools and differences 
        else:
            #PI has been found 
            if len(arr[i])==4:
                del(arr[i][-1])
                del(arr[i][-1])
            elif len(arr[i]) == 2:
                del(arr[i][-1])
            elif len(arr[i])== 7:
                del(arr[i][-1])
                del(arr[i][-1])
                del(arr[i][-1])
        new_arr.append(arr[i])
        
    # This cleans up the empty lists inside the list 
    try:
        while True:
            new_arr.remove([])
    except ValueError:
        pass
    new_new_arr = []
    
    #This sorts the cubes and deletes duplicates
    for i in range(len(new_arr)):
        new_arr[i].sort()
    [new_new_arr.append(x) for x in new_arr if x not in new_new_arr]
    
    #retuns the array that includes the PIs
    return new_new_arr
            
```

```python
def read_input():
    #csv file name 
    filename = "C:\\Users\\Thalia Lightstone\\Documents\\EE-26\\ee26qm.txt"
    
    #initializing titles and rows 
    minterms = []
    dont_cares = []
    
    #reading txt file 
    with open (filename) as f:
        for line in f:
            get_line = line.split('+')
            
            #miterms
            min_term = get_line[0][2:-1]
            list_min = list(min_term.split(','))
            last_string = list_min[-1]
            last_char = list(last_string)
            
            if (last_char[-1] ==')'):
                del(last_char[-1])
                seperator = ''
                last_string = seperator.join(last_char)
            
    
            list_min[-1] = last_string
            int_min = [int(x) for x in list_min]
                
            minterms.append(int_min)

            
            #dont cares 
            if len(get_line) > 1: 
                dont_care = get_line[1][2:-1]
                list_dc = list(dont_care.split(','))
                last_dc = list_dc[-1]
                dc_char = list(last_dc)
                
                if (dc_char[-1] ==')'):
                    del(dc_char[-1])
                    seperator = ''
                    last_dc = seperator.join(dc_char)
                
                list_dc[-1] = last_dc
                int_dc = [int(x) for x in list_dc]
                dont_cares.append(int_dc)
            else:
                dont_cares.append([])
        return(minterms,dont_cares)
            


    
minterms=[]
dont_cares=[]
(minterms,dont_cares) = read_input()
print(minterms)
print(dont_cares)
```

```python
def mintermsToPIs(minterms,dont_cares):
    print("The minterms are: ", minterms)
    print("The don't cares are ", dont_cares)
    
    # number of 1s bits per integer 
    zero_bit = [0]
    one_bit = [1,2,4,8]
    two_bit = [3,5,6,9,10,12]
    three_bit = [7,11,13,14]
    four_bit = [15]
    
    #combine minterms and dont cares
    
    
    
    # Orders the minterms into arrays in the 0's column of the chart
    m = set(minterms).union(set(dont_cares))
    n_0_bit = [x for x in zero_bit if x in m]
    n_1_bit = [x for x in one_bit if x in m]
    n_2_bit = [x for x in two_bit if x in m]
    n_3_bit = [x for x in three_bit if x in m]
    n_4_bit = [x for x in four_bit if x in m]
    
    #adds bools to the end of the 0's cube column
    n_0_bit = addBool(n_0_bit)
    n_1_bit= addBool(n_1_bit)
    n_2_bit= addBool(n_2_bit)
    n_3_bit= addBool(n_3_bit)
    n_4_bit= addBool(n_4_bit)
    
    # finds all the one's cubes 
    onescube0,n_0_bit,n_1_bit= findOnesCube(n_0_bit,n_1_bit)
    onescube1,n_1_bit,n_2_bit= findOnesCube(n_1_bit,n_2_bit)
    onescube2,n_2_bit,n_3_bit= findOnesCube(n_2_bit,n_3_bit)
    onescube3,n_3_bit,n_4_bit= findOnesCube(n_3_bit,n_4_bit)
    
    
    # finds all the two's cubes
    twocube0,onescube0,onescube1 = findTwosCube(onescube0,onescube1)
    twocube1,onescube1,onescube2 = findTwosCube(onescube1,onescube2)
    twocube2,onescube2,onescube3 = findTwosCube(onescube2,onescube3)
    
    print("0's True means it has been checked off: ")
    print(n_0_bit)
    print(n_1_bit)
    print(n_2_bit)
    print(n_3_bit)
    print(n_4_bit)
    
    print("ones cubes")
    print(onescube0)
    print(onescube1)
    print(onescube2)
    print(onescube3)
    
    print("twos cube")
    print(twocube1)
   
    
    unuseds = []
    # adds all of the cubes into one list so that we can find the Essential PIs
    all_lists = n_0_bit + n_1_bit + n_2_bit + n_3_bit + n_4_bit + onescube0 + onescube1 + onescube2 + onescube3 + twocube1
    unuseds = findUnuseds(all_lists)

    print("Prime Implicants are: ",unuseds)
    print("Now looking for Essential PIs")
    
    return unuseds
    
unuseds = mintermsToPIs(minterms[0],dont_cares[0])
```

```python
def findPIs(u,m):
#This function finds the PIs
    table = m.copy()
    unu = u.copy()
    for i in range(len(table)):
        table[i] = [table[i],0]
    for i in range(len(table)):
        for j in range(len(unu)):
            for k in range(len(unu[j])):
                if table[i][0] == unu[j][k]:
                    table[i][1] += 1
                    table[i].append(j)
    #print("newminterms",table)
    
    for i in range(len(table)):
        if table[i][1] == 1: 
            unu[table[i][2]].append(True)
    
    for i in range(len(unu)):
        for j in range(len(unu[i])):
            if unu[i][-1] != True and unu[i][-1] != False :
                unu[i].append(False)
    
    print("PIs", unu)
    return unu,table
    
PIs, mintermsTable = findPIs(unuseds,minterms[0])
```

```python
def findSOP(EPIs):
#Using the essential PIs we will turn it into SOP
    size = 0
    for i in range(len(PIs)):
        if PIs[i][-1]==True:
            size += 1
            
    binary = []
    for i in range (size):
        binary.append([])
    numPIs = 0

    for i in range(len(PIs)):
        if PIs[i][-1] == True:
            #essential PI found
            for j in range(len(PIs[i])-1):
                temp_bin = PIs[i][j]
                temp_bin = '{:04b}'.format(temp_bin)
                binary[numPIs].append((temp_bin))

            numPIs+=1
    print("These are the essential PIs in binary form:",binary)
    
    #turns the values into binary form
    sop =[]

    newsop = []
    for i in range(len(binary)):
        if len(binary[i])-1 > 0:
            for j in range(len(binary[i])-1): 
                for k in range(4):
                    sop.append('0')
                    sop[k] = binary[i][0][k]
                    if ((binary[i][0][k] != binary[i][j+1][k])):
                        sop[k] = 'x'

        else: 
            for k in range(4):
                sop.append('0')
                sop[k]=(binary[i][0][k])
        del sop[4:]            
        newsop.append( ''.join([str(x) for x in sop]) )
        sop = []
    print("Turning into Sum Of Products with don't cares", newsop)
    
    #into SOP
    new_newsop = []
    answer = []
    for i in range(len(newsop)):
        if newsop[i][0] == '0':
            new_newsop.append('A\'')
        if newsop[i][0] == '1':
            new_newsop.append('A')
        if newsop[i][1] == '0':
            new_newsop.append('B\'')
        if newsop[i][1] == '1':
            new_newsop.append('B')
        if newsop[i][2] == '0':
            new_newsop.append('C\'')
        if newsop[i][2] == '1':
            new_newsop.append('C')
        if newsop[i][3] == '0':
            new_newsop.append('D\'')
        if newsop[i][3] == '1':
            new_newsop.append('D')
        SOP = ''.join(new_newsop)
        answer.append(SOP)
        SOP = ''
        new_newsop = []
                
    print("The Sum of Products is", answer)
    
findSOP(PIs)
```
